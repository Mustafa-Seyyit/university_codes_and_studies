# Doğrusal Regresyon

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("linear-regression-dataset.csv", sep=";")
```

```python
df.head()
```

```python
print(df.columns)
```

```python
df.rename(columns={
        "experience":"deneyim",
        "salary":"maas"}, inplace=True)
```

```python
plt.scatter(df.deneyim,df.maas)
plt.xlabel("Deneyim")
plt.ylabel("Maas")
plt.show()
```

```python
x= df.deneyim.values
```

```python
x.shape
```

```python
x=df.deneyim.values.reshape(-1,1)
y=df.maas.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
linear_reg.fit(x,y)
```

```python
#prediction
b0 = linear_reg.predict([[0]])
print("b0: ",b0)
```

```python
b0_ = linear_reg.intercept_
print("b0: ",b0_)
```

```python
b1 = linear_reg.coef_
print("b1: ", b1)
```

```python
new_salary = 1663 + 1138*11
print(new_salary)
```

```python
b11 = linear_reg.predict([[11]])
print("b11: ", b11)
```

```python
x_sorted = np.sort(x, axis=0)
plt.plot(x_sorted, linear_reg.predict(x_sorted), color="red")
plt.scatter(x, y)
plt.show()
```

```python
from sklearn.metrics import r2_score
print("r^2 score: ", r2_score(y, linear_reg.predict(x)))
```
