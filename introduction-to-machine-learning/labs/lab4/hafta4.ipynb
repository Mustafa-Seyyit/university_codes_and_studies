```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("../lab4/breast-cancer.csv")
data.head()
```

```python
data.info()
```

```python
print(data.columns)
```

```python
data.drop(["id"], axis=1, inplace=True)
data.tail()
```

```python
M = data[data.diagnosis=="M"] # M = malinant - kötü huylu tümör
B = data[data.diagnosis=="B"] # B = benign - iyi huylu tümör
```

```python
#scatter plot
plt.scatter(M.radius_mean, M.texture_mean, color="red", label="kötü huylu", alpha=0.6)
plt.scatter(B.radius_mean, B.texture_mean, color="green", label="iyi huylu", alpha=0.6)

plt.xlabel("radius_mean") # x ekseni, tümörün ortalama yarıçapını gösterir.
plt.ylabel("texture_mean") # y ekseni, tümör dokusunun ortalama değerini gösterir.

plt.legend()
plt.show()
```

```python
data.diagnosis = [1 if each == "M" else 0 for each in data.diagnosis]
data.diagnosis
```

```python
y = data.diagnosis.values
x_data = data.drop(["diagnosis"], axis=1)
```

```python
print(y[:30])
x_data.head(2)
```

```python
#normalization
x = (x_data -np.min(x_data)) / (np.max(x_data) - np.min(x_data))
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
# KNN modeli
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 3) # n_neighbors = k
knn.fit(x_train, y_train)
```

```python
prediction = knn.predict(x_test)
prediction
```

```python
print("{} nn score: {}".format(3, knn.score(x_test, y_test)))
```

```python
# En iyi performansı veren K değerini bulalım.

score_list = []

for each in range(1,15):
    knn2 = KNeighborsClassifier(n_neighbors = each)
    knn2.fit(x_train, y_train)
    score_list.append(knn2.score(x_test, y_test))

best_k = score_list.index(max(score_list)) + 1
best_accuracy = max(score_list)

plt.figure(figsize =(8,5))

plt.plot(range(1, 15), score_list,
        marker="o",
        color="blue",
        linewidth=2)

plt.xlabel("K Değeri", fontsize=12)
plt.ylabel("Accuracy", fontsize=12)
plt.title("K Değerine Göre Model Başarısı", fontsize=14)

plt.grid(True)

plt.axvline(best_k, color="red", linestyle="--", label=f"En iyi K = {best_k}")

plt.legend()
plt.show()

print("Best accuracy is {:.4f} with k = {}".format(best_accuracy, best_k))
```

```python
error = []

for i in range(1, 40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)

    pred_i = knn.predict(x_test)

    # Ortalama hata oranı
    error.append(np.mean(pred_i != y_test))


# Grafik
plt.figure(figsize=(12,6))

plt.plot(range(1,40),
        error,
        color="red",
        linestyle="dashed",
        marker="o",
        markerfacecolor="blue",
        markersize=8)

plt.title("Error Rate vs K Value")
plt.xlabel("K Value")
plt.ylabel("Mean Error")

plt.grid(True)

plt.show()

# En düşük hata oranını bul
best_k = error.index(min(error)) + 1
best_error = min(error)

print("En düşük hata oranı:", best_error)
print("En iyi K değeri:", best_k)
```

# Çapraz Doğrulama (Cross Validation)

```python
best_k = score_list.index(max(score_list)) + 1

knn_best = KNeighborsClassifier(n_neighbors=best_k)
knn_best.fit(x_train, y_train)

y_pred = knn_best.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

#confusion matrix hesapla
cm = confusion_matrix(y_test, y_pred)

print("Confusion Matrix:\n", cm)

#grafik olarak göster
plt.figure(figsize=(6,4))
sns.heatmap(cm,
            annot=True,
            fmt="d",
            cmap="Blues",
            xticklabels=["Benign (B)", "Malignant (M)"],
            yticklabels=["Beign (B)", "Malignant (M)"])

plt.xlabel("Tahmin Edilen Sınıf")
plt.ylabel("Gerçek Sınıf")
plt.title("KNN Confusion Matrix")

plt.show()
```
