```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("Social_Network_Ads.csv")

data.head()
```

```python
plt.figure(figsize=(8, 6), dpi=600)

scatter = plt.scatter(data['Age'],
    data['EstimatedSalary'],
    c=data['Purchased'],
    cmap='rainbow'
)

plt.xlabel("Yaş")
plt.ylabel("Tahmini Maaş")
plt.title("Kullanıcıların Satın Alım Davranışları")

handles, _ = scatter.legend_elements()
labels = ["Satın Almadı", "Satın Aldı"]

plt.legend(handles, labels, title="Satın Alım Durumu")

plt.show()
```

```python
# Feature ve Target Seçimi

X = data.iloc[:, [2, 3]].values
y = data.iloc[:, 4].values
```

```python
# Eğitim ve Test verilerini ayarlama
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(criterion="gini", random_state=42)
model.fit(X_train, y_train)
```

```python
# Tahmin ve değerlendirme
from sklearn.metrics import confusion_matrix, accuracy_score

y_pred = model.predict(X_test)

cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)

print("Accuracy: ", accuracy_score(y_test, y_pred))
```

```python
import seaborn as sns

plt.figure(figsize=(6,5), dpi=600)

sns.heatmap(
    cm,
    annot=True,  # hücre içine sayı yaz
    xticklabels=["Satın Almadı","Satın Aldı"],
    yticklabels=["Satın Almadı","Satın Aldı"]
)

plt.xlabel("Tahmin Edilen")
plt.ylabel("Gerçek Değer")
plt.title("Confusion Matrix")

plt.show()
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(10,6), dpi=600)

plot_tree(
    model,
    feature_names = ["Age", "Salary"],
    class_names = ["Satın Almadı", "Satın Aldı"],
    filled=True,
    max_depth=2 #sadece ilk 2 seviye
)

plt.show()
```
