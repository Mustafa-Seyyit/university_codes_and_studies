```python
import pandas as pd
data = {
    'Gender' :  ['m','f','f','m','f','m','m'],
    'Height' :  [172, 171, 169, 173, 170, 175, 178]
}
df_sample = pd.DataFrame(data)
f_filter = df_sample['Gender'] == 'f'
print(df_sample[f_filter]) #kadın verilerini görüntüleme
```

```python
m_filter = df_sample['Gender'] == 'm'
print(df_sample[m_filter])  #erkek verileni görüntüleme
```

```python
f_avg= df_sample[f_filter]['Height'].mean()
m_avg= df_sample[m_filter]['Height'].mean()
df_output = pd.DataFrame({'Gender':['f','m'],'Height':[f_avg, m_avg]})
print(df_output) #uzun yoldan ortalama bulma
```

```python
df_sample.groupby('Gender').mean() #kısa yokdan groupby ile ortalama bulma
```
