```python
import pandas as pd

data = {
    "calories" : [420,380,390],
    "duration" : [50, 40, 45]
}
#load data into a DataFrame Object:
df = pd.DataFrame(data)
print(df)
#pandas'ın DataFrame metodu hakkında(2 boyutlu matris yani tablo oluşturur)
```

```python
print(df.loc[0]) #df.loc[?] = belirli bir satırdaki döndürür
```

```python
print(df.loc[[0, 1]]) # df.loc[[?, ?]] dönmesini istediğin satırları döndürür
```

```python
df = pd.DataFrame(data, index=["day1", "day2", "day3"])
#bu ifade satırları isimlendirmemizi sağlar
print(df)
```

```python
print(df.loc["day3"])
#girilen satırın bilgilerini gösterir
```
