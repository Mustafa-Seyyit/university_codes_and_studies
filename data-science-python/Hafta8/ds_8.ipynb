```python
filename = "ds.txt"
def write_to(message):
    with open(filename,"a+") as file:
        file.write(message+"\n")
```

```python
write_to("merhabalar")
write_to("ben mustafa")
with open(filename,"r") as file:
    lines = file.readlines()
    for line in lines:
        print(line)
```

```python
import os 
if os.path.exists(filename):
    write_to("dosya var")

os.remove(filename)
```
